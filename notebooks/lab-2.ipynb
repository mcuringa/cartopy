```python
# import pandas library
import pandas as pd

# read the data from a URL
url = "https://raw.githubusercontent.com/mcuringa/cartopy/main/notebooks/data/large-us-cities.csv"
url = "data/large-us-cities.csv"

# read the data into a pandas DataFrame
df = pd.read_csv(url)

# display the first few rows of the DataFrame
df
```

```python
# select just some rows from our table

# let's get just cities in California
df[df['state'] == 'CA']
```

```python
# combine terms
# get CA cities with population greater than 250,000 in 2022
df[(df['state'] == 'CA') & (df['pop_2022'] > 250000)]
```

```python
# use isin() to select rows with values in a list
# create a DataFrame with just New England
new_england = df[df['state'].isin(['CT', 'ME', 'MA', 'NH', 'RI', 'VT'])]
# get just the columns we care about
new_england = new_england[["name", "state", "pop_2022"]]
# rename pop_2022 to population
# new_england = new_england.rename(columns={"pop_2022": "population"})
new_england.pop_2022 = new_england.pop_2022.apply(lambda x: f"{x:,}")
display(new_england)
df
```
