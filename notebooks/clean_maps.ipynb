```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.colors import to_hex

import os
path = '/home/mxc/Projects/cartopy/_data'

t = []
for root, dir, files in os.walk(path):
    t.extend([os.path.join(root, f) for f in files if f.endswith("kml") ])

t
```

```python

runs = []
for i, f in enumerate(t):
    folder = f.split("/")[-2]
    runner = folder.split("_")[0]
    df = gpd.read_file(f)
    df["type"] = "route"
    df.iloc[1, df.columns.get_loc('type')] = 'start'
    df.iloc[2, df.columns.get_loc('type')] = 'end'
    df["runner"] = runner 
    df["index"] = i
    runs.append(df)

data = pd.concat(runs)
data["color"] = "orange"

data
```

```python

cmap = plt.get_cmap('tab20')
data["color"] = data["index"].apply(lambda x: to_hex(cmap(x)))
# data.explore(color=data["color"], tooltip="runner", legend=True)

data = data[['index',  'runner',  'color', 'type', 'geometry']].copy()
data.to_file("runs.geojson", driver='GeoJSON')
data
```
