```python
import geopandas as gpd
campus = gpd.read_file("data/campus.geojson")
campus = campus[~campus.key.isnull()]
campus.id = range(1, len(campus) + 1)  
campus.to_file("data/campus.geojson", driver="GeoJSON")
```

```python
campus = gpd.read_file("data/campus.geojson")
m = campus.explore(color="violet", fill_opacity=1)
data = gpd.read_file("data/run.kml")
data.iloc[[0]].explore(m=m, color="blue")
data.iloc[1:].explore(m=m, color='red', marker_kwds={"radius": 10})
```
